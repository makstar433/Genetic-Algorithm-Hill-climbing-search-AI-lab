```python
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
x=np.linspace(-2,2,100)
y=np.linspace(-2,2,100)
plt.plot(x,y)
X,Y=np.meshgrid(x,y)
#z=eval("np.exp(-(X**2+Y**2))")
z=eval("(1-X)**2+100*(Y-X**2)**2")
ax.plot_surface(X,Y,z)
```
