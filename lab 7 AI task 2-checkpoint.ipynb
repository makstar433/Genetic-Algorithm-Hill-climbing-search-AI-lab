```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Define GA parameters
genotype_size = 32
population_size = 100
max_generations = 1000
tournament_size = 3
crossover_rate = 0.8
mutation_rate = 0.01

def encode_genotype(x, y):
    x = int((x + 2) * 2**16 / 4)
    y = int((y + 2) * 2**16 / 4)
    return f"{x:016b}{y:016b}"

def decode_genotype(genotype):
    x = int(genotype[:16], 2) * 4 / 2**16 - 2
    y = int(genotype[16:], 2) * 4 / 2**16 - 2
    return x, y

# Function 1
def fitness_func1(individual):
    x, y = decode_genotype(individual)
    return -(1-x)**2 - 100*(y-x**2)**2

# Function 2
def fitness_func2(individual):
    x, y = decode_genotype(individual)
    return np.exp(-(x**2 + y**2))

def crossover(parent1, parent2, crossover_rate):
    child1 = parent1
    child2 = parent2
    if np.random.rand() < crossover_rate:
        crossover_point = np.random.randint(1, genotype_size - 1)
        child1 = parent1[:crossover_point] + parent2[crossover_point:]
        child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

def mutation(individual, mutation_rate):
    mutated_individual = ""
    for bit in individual:
        if np.random.rand() < mutation_rate:
            mutated_individual += "0" if bit == "1" else "1"
        else:
            mutated_individual += bit
    return mutated_individual

def tournament_selection(population, fitness_func, tournament_size):
    tournament = np.random.choice(population, size=tournament_size)
    tournament_fitness = np.array([fitness_func(individual) for individual in tournament])
    return tournament[np.argmax(tournament_fitness)]

def initialize_population(population_size):
    return ["".join(np.random.choice(["0", "1"], size=genotype_size)) for i in range(population_size)]

def run_ga(fitness_func, genotype_size, population_size, max_generations, tournament_size, crossover_rate, mutation_rate):
    population = initialize_population(population_size)
    best_fitness = -np.inf
    best_individual = None
    for i in range(max_generations):
        fitness_values = np.array([fitness_func(individual) for individual in population])
        best_index = np.argmax(fitness_values)
        if fitness_values[best_index] > best_fitness:
            best_fitness = fitness_values[best_index]
            best_individual = population[best_index]
            print(f"Generation {i}, Best Fitness: {best_fitness}")
        new_population = [best_individual]
        while len(new_population) < population_size:
            parent1 = tournament_selection(population, fitness_func, tournament_size)
            parent2 = tournament_selection(population, fitness_func, tournament_size)
            child1, child2 = crossover(parent1, parent2, crossover_rate)
            child1 = mutation(child1, mutation_rate)
            child2 = mutation(child2, mutation_rate)
            new_population.append(child1)
            if len(new_population) < population_size:
                            new_population.append(child2)
        population = new_population
    return decode_genotype(best_individual)

# Function 1
print("Running GA for Function 1...")
best_solution1 = run_ga(fitness_func1, genotype_size, population_size, max_generations, tournament_size, crossover_rate, mutation_rate)
print("Best solution for Function 1:", best_solution1)

# Function 2
print("Running GA for Function 2...")
best_solution2 = run_ga(fitness_func2, genotype_size, population_size, max_generations, tournament_size, crossover_rate, mutation_rate)
print("Best solution for Function 2:", best_solution2)

# Plot the results
fig = plt.figure(figsize=(12, 6))

# Function 1 plot
ax = fig.add_subplot(121, projection='3d')
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = (1 - X)**2 + 100*(Y - X**2)**2
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.set_title("Function 1")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.scatter(best_solution1[0], best_solution1[1], fitness_func1(encode_genotype(*best_solution1)), s=200, c='r')

# Function 2 plot
ax = fig.add_subplot(122, projection='3d')
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='plasma', alpha=0.8)
ax.set_title("Function 2")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.scatter(best_solution2[0], best_solution2[1], fitness_func2(encode_genotype(*best_solution2)), s=200, c='r')

plt.show()
```
